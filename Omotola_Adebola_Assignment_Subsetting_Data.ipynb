## Boolean Indexing

**7. CHECK if certain values exist with the `isin` method**

This method can be called on a pandas series or dataframe object to check if certain values are present or not. By default, It returns a boolean dataframe or series with True or False, depending on whether the specified values exist or not.<br>

Lets read in the `healthexp` dataset

```python
# read it below...
healthexp = sns.load_dataset('healthexp')
print(healthexp.shape)
healthexp.head()
```

```python
# Inspect the first 5 observations...
healthexp.head()
```

```python
# Use isin to get all records from the healthexp Dataframe for the years 2010, 2015, 2020
# List of years you want to filter
years_to_filter = [2010, 2015, 2020]

# Use .isin() to filter records for the specified years
filtered_healthexp = healthexp[healthexp['Year'].isin(years_to_filter)]

# Display the filtered DataFrame
print(filtered_healthexp)
```

```python
# Next, let's use the isin method to get the number of times that 2010 occur in the year column

# Use .isin() to create a boolean mask where 2010 occurs
is_2010 = healthexp['Year'].isin([2010])

# Use .sum() to count the number of True values (i.e., the occurrences of 2010)
count_2010 = is_2010.sum()

# Display the count
print("Number of occurrences of 2010:", count_2010)
```

**8. SUBSETTING data based on Boolean operator `==`...**

This method can be called on a pandas series or dataframe to subset data based on a condition. If the condition were True, then the subset is retrieved, if not null.

```python
# Now, let's subset all data for when the Country is France in healthexp
# Create a boolean mask where the 'Country' column is equal to 'France'
is_france = healthexp['Country'] == 'France'

# Use the boolean mask to subset the DataFrame
france_data = healthexp[is_france]

# Display the subsetted DataFrame
print(france_data)
```

**9. SUBSETTING data based on Boolean operators `&` and `|`...**

This method can be called on a pandas series or dataframe to subset data. Note that `&` will only return data when the two conditions are True, while `|` will return data if either condition is True as well as if both conditions are True.

```python
# Now let's select data for when Life_Expectancy > 70 and Year < 2000
# Create a boolean mask based on the conditions
mask = (healthexp['Life_Expectancy'] > 70) & (healthexp['Year'] < 2000)

# Use the boolean mask to subset the DataFrame
selected_data = healthexp[mask]

# Display the subsetted DataFrame
print(selected_data)
```

```python
# Now let's select data for Great Britain in the Year 2020
# Create a boolean mask based on the conditions
mask = (healthexp['Country'] == 'Great Britain' ) & (healthexp['Year']==2000)

# Use the boolean mask to subset the DataFrame
selected_data = healthexp[mask]

# Display the subsetted DataFrame
print(selected_data)


```

```python
# Now let's select data for Great Britain in the Year 1975
mask = (healthexp['Country'] == 'Great Britain' ) & (healthexp['Year']==1975)

# Use the boolean mask to subset the DataFrame
selected_data = healthexp[mask]

# Display the subsetted DataFrame
print(selected_data)
```

**10. SUBSETTING data based on the Filter Method...**

An alternative method to select columns is with the filter method. This method is flexible and searches column names (or index labels) based on which parameter is used... `<like, items, regex>`.

Let's read the `anagrams` dataset

```python
# Read it here...
anagrams = sns.load_dataset('anagrams')
print(anagrams.shape)
anagrams.head()
```

```python
# Inspect the first 5 observations...
anagrams.head()
```

```python
# Let's use the like parameter to find columns that have the word `num` in them
filtered_columns = anagrams.filter(like='num')

# Display the columns that contain 'num'
print(filtered_columns.columns)
```

```python
# Now, let's use the items parameter to find specific columns.
# The good thing about items is that if a column don't exist, it won't throw an error
# so select 2 existing columns and one fictional column within the items parameter...

# Assuming your DataFrame is named df
# Specify the columns you want to select, including one fictional column
selected_columns = ['num1', 'num2']

# Use the items parameter to select these columns
selected_data = anagrams[selected_columns]

# Display the selected data
print(selected_data)

```
